```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://docs.google.com/spreadsheets/d/1VMaw2oCn0ABitd-alLAEsEhGS1Je2UFNLu76TKrIH7w/gviz/tq?tqx=out:csv&sheet=Raw_Medical_Data_for_day1'
df = pd.read_csv(url, index_col = 0)
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax = ax, stat='probability')
```

```python
mean = df['Income'].mean()
mean
```

```python
fig,ax= plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax = ax, stat='probability')
ax.axvline(mean);
```

```python
#change the color of the line to red
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax = ax, stat='probability')
ax.axvline(mean, color = 'red')
```

```python
#add legend and title
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax=ax, stat='probability')
ax.axvline(mean, color = 'red', label='Mean')
ax.set_title('Income')
ax.legend();
```

```python
#add value for mean with an f string
fig, ax= plt.subplots(figsize=(10,6))
sns.histplot(df['Income'], ax=ax, stat='probability')
ax.axvline(mean,color='red', label=f'Mean = {mean:.2f}')
ax.set_title('Income')
ax.legend();
```

```python
#assign column name a variable
col = 'Income'
```

```python
col = 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature, ax=ax,stat='probability')
ax.axvline(mean, color='red', label=f'Mean = mean = {mean:,.2f}')
ax.axvline(median, ls='--', color='green', label=f'Median = {median:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
#Next we want to highlight the values that correspond to the range between -1 std and +1 std from the mean.
#First define +1 and -1 standard deviation from the mean.  
#You can graph both of these lines.
col= 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax= plt.subplots(figsize=(10,6))
sns.histplot(feature, ax=ax, stat='probability')
ax.axvline(plus_one_std, color = 'red', label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label=f'-1 std = {minus_one_std:,.2f}')
ax.set_title(f'{col}')
ax.legend();
```

```python
#When trying to highlight the space between two lines, you can use ax.axvspan()
col= 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax= plt.subplots(figsize=(10,6))
sns.histplot(feature, ax=ax, stat='probability')
ax.axvline(plus_one_std, color = 'red', label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label=f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow')
ax.set_title(f'{col}')
ax.legend();
```

```python
#It would be better if the yellow area were behind the rest of the plot.  We can do this by specifying "zorder = 0".
col= 'Income'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax= plt.subplots(figsize=(10,6))
sns.histplot(feature, ax=ax, stat='probability')
ax.axvline(plus_one_std, color = 'red', label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label=f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder= 0)
ax.set_title(f'{col}')
ax.legend();
```
